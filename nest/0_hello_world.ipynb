```python
from matplotlib import pyplot as plt
import nest
import nest.voltage_trace
nest.ResetKernel()
```

```python
neuron = nest.Create('iaf_psc_exp')  # create neuron
spikegenerator = nest.Create('spike_generator')  # create spike generator
spikegenerator.spike_times = [10., 50.]  # set the spike times
voltmeter = nest.Create('voltmeter')  # create voltmeter
nest.Connect(spikegenerator, neuron, syn_spec={'weight': 1e3})
nest.Connect(voltmeter, neuron)  
nest.Simulate(100.)  # simulate 100 ms
```

```python
%matplotlib inline
nest.voltage_trace.from_device(voltmeter)
```
